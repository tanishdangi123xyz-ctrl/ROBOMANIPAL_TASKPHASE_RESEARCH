ENCODING CATEGORICAL DATA:

USING NUMERICAL ORDER MIGHT MISLEAD THE MACHINE LEARNING MODEL TO THINK THE LABELS/NAMES INFLUENCE THE OUTCOME (MISINTERPRETED CORELATION)

ONE HOT ENCODING:

CONVERTS EACH INTO DIFFERENT COLUMNS AND USES VECTORS SUCH AS [1,0,0], [0,1,0] AND [0,0,1]. SO NOW THERE IS NO NUMERICAL ORDER IN THE DATA AND ONLY 0s AND 1s ARE USED. (USEFUL WHEN CATEGORICAL VARIABLES)


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

features_to_drop = ['Formatted Date', 'Temperature (C)', 'Daily Summary']
X_df = df.drop(columns=features_to_drop)
X = X_df.values
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False), [0, 1])], remainder='passthrough')
X_encoded = np.array(ct.fit_transform(X))
```
